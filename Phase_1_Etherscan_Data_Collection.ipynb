```python
import requests
import pandas as pd

# Replace with your actual Etherscan API Key
ETHERSCAN_API_KEY = "MM1P5U7G4WUKUXBR57DXUCSTS24CITESVK"

# Replace with the wallet address you want to analyze
wallet_address = "0x81bb35c4152B605574BAbD320f8EABE2871CE8C6"

# Define the API URL
API_URL = f"https://api.etherscan.io/api?module=account&action=txlist&address=0x81bb35c4152B605574BAbD320f8EABE2871CE8C6&startblock=0&endblock=99999999&sort=desc&apikey=MM1P5U7G4WUKUXBR57DXUCSTS24CITESVK"

# Make the API request
response = requests.get(API_URL)
data = response.json()

# Check if the request was successful
if data["status"] == "1":
    transactions = data["result"]

    # Convert to a DataFrame for better readability
    df = pd.DataFrame(transactions)

    # Select relevant columns
    df = df[['hash', 'from', 'to', 'value', 'gas', 'gasPrice', 'timeStamp']]

    # Convert timestamps to readable dates
    df["timeStamp"] = pd.to_datetime(df["timeStamp"], unit='s')

    # Convert value from Wei to ETH
    df["value"] = df["value"].astype(float) / (10**18)

    print(df.head())  # Show first few transactions
else:
    print("Error fetching data:", data["message"])

```
